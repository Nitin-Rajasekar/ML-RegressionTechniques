```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
X, y =  make_blobs(n_samples=100, centers=[[2,4],[4,2]], random_state=20)
```

```python
def sigmoid(w,X):
    sig=1.0/(1+np.exp(-1*(w[0]+w[1]*X[0]+w[2]*X[1])))
    return sig
```

```python
def loss_func(w):
    count=0
    loss=0
    for i in X:
        expected=y[count]
        loss-=y[count]*np.log(sigmoid(w,i))+(1-y[count])*np.log(1-sigmoid(w,i))
        count=count+1
    
    loss=loss/count
    return loss
```

```python
w=[1,1,1]
learning_rate=0.1

for i in range(1000):
    w_0_new=[w[0]+0.01,w[1],w[2]]
    derivative_0=(loss_func(w_0_new)-loss_func(w))/0.01
    w[0]=w[0] - learning_rate*derivative_0
    
    w_1_new=[w[0],w[1]+0.01,w[2]]
    derivative_1=(loss_func(w_1_new)-loss_func(w))/0.01
    w[1]=w[1] - learning_rate*derivative_1
    
    w_2_new=[w[0],w[1],w[2]+0.01]
    derivative_2=(loss_func(w_2_new)-loss_func(w))/0.01
    w[2]=w[2] - learning_rate*derivative_2
```

```python
plt.figure(figsize=(10,10))

#Visualize dataset 
plt.plot(X[:,0][y==0],X[:,1][y==0],'o',color='red')
plt.plot(X[:,0][y==1],X[:,1][y==1],'^',color='blue')


x_points=np.linspace(0,6)
y_points=(-1*(w[0]+w[1]*x_points))/w[2]

plt.plot(x_points, y_points)
```

Logistic regression made use of in classification analysis to estimate the likelihood of a certain event occurring., or more specifically the probability of a certain data point belonging to a given class. It may thus be looked at as a supervised classification algorithm.
<br>
It makes use of a computed derived threshold to predict the probability of a data entry belonging to the category in question.
<br>
To do this, similar to linear regression, it uses a loss function to compute the equation of a function which would fit best as a decision boundary using gradient descent. 
<br>
It then makes use of the said decision boundary and activation function to assign classes to the data points.

As can be seen in the above loss function, 

The loss function used is the summation of

$y.log(p(y))+(1-y).log(1-p(y))$

for every class y, divided by the total number of class $N$.


Mean square error cannot be used a substiture loss function in logistic regression like linear regression.
<br>
The loss function in logistic regression must be convex. This is because in a classification algorithm, the loss function must increase similarly if the classifier is biased either way, reaching an ideal value in the middle when both classification towards both classes are probable. Since mean sqaure error can never be a convex function, it cannot be used.
